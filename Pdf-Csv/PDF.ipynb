```python
import PyPDF2
```

```python
f = open('Working_Business_Proposal.pdf', 'rb')
```

```python
f.read()
```

```python
pdf_reader = PyPDF2.PdfFileReader(f)
```

```python
pdf_reader.numPages
```

```python
page_one = pdf_reader.getPage(0)
```

```python
page_one_text = page_one.extractText(page_one)
```

```python
page_one_text
```

```python
pdf_writer = PyPDF2.PdfFileWriter()
```

```python
pdf_writer.addPage(page_one)
```

```python
pdf_output = open('Grabbing_new_page.pdf', 'wb')
```

```python
pdf_writer.write(pdf_output)
```

```python
f.close()
```

```python
pdf_output.close()
```
